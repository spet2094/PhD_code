```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
input_path= "/Users/sophieayling/Library/CloudStorage/GoogleDrive-sophie2ayling@gmail.com/My Drive/PhD/06_Data and Modelling/thesis_data/combined_estimates/"
output_path= "/Users/sophieayling/Library/CloudStorage/GoogleDrive-sophie2ayling@gmail.com/My Drive/PhD/06_Data and Modelling/thesis_data/mobility/ward_level/plots/"
```

```python
data=pd.read_csv(input_path+"comb_761_wards_indiv_nld_ld.csv")
data.head()
```

```python
plt.figure(figsize=(10, 6))

# Plot each histogram with different colors
sns.histplot(data['excl_moves_from_source_eld'], color='red', kde=False, label='Early LD', linewidth=2, alpha=0.2, edgecolor ='red')
sns.histplot(data['excl_moves_from_source_lld'], color='blue', kde=False, label='Late LD', linewidth=2, alpha=0.2, edgecolor ='blue')
sns.histplot(data['excl_moves_from_source_nld'], color='green', kde=False, label='Non LD', linewidth=2, alpha=0.2, edgecolor ='green')

# Set axis limits
plt.xlim(0, 200000)


# Add legend
plt.legend(title="Legend", fontsize='small', loc='upper right')

# Add title
plt.title("Outflow Count Non Lockdown, Early Lockdown & Late Lockdown", fontsize=12)

# Manually set x-axis label
plt.xlabel("Outflow counts")


# Show the graph
plt.savefig(output_path+'flow_count_nld_eld_lld.png', dpi=300, facecolor = 'white')
```

```python
plt.figure(figsize=(10, 6))

# Plot each histogram with different colors
sns.histplot(data['excl_trips_per_subscriber_eld'], color='red', kde=False, label='Early LD', linewidth=1, alpha=0.2, edgecolor ='red', binwidth = .25)
sns.histplot(data['excl_trips_per_subscriber_nld'], color='blue', kde=False, label='Late LD', linewidth=1, alpha=0.2, edgecolor ='blue', binwidth = .25)
sns.histplot(data['excl_trips_per_subscriber_lld'], color='green', kde=False, label='Non LD', linewidth=1, alpha=0.2, edgecolor ='green', binwidth = .25)

# Set axis limits
#plt.xlim(0, 200000)


# Add legend
plt.legend(title="Legend", fontsize='small', loc='upper right')

# Add title
plt.title("Outflow Count Per Subscriber Non Lockdown, Early Lockdown & Late Lockdown", fontsize=12)
# Manually set x-axis label
plt.xlabel("Outflows per subscriber")


# Show the graph
plt.savefig(output_path+'flow_count_per_sub_nld_eld_lld.png', dpi=300, facecolor = 'white')
```

```python
## check out subscriber count distributions 

plt.figure(figsize=(10, 6))

# Plot each histogram with different colors
sns.histplot(data['eld_subscriber_count'], color='red', kde=False, label='Early LD', linewidth=2, alpha=0.2, edgecolor ='red')
sns.histplot(data['nld_subscriber_count'], color='blue', kde=False, label='Late LD', linewidth=2, alpha=0.2, edgecolor ='blue')
sns.histplot(data['lld_subscriber_count'], color='green', kde=False, label='Non LD', linewidth=2, alpha=0.2, edgecolor ='green')

# Set axis limits
plt.xlim(0, 200000)


# Add legend
plt.legend(title="Legend", fontsize='small', loc='upper right')

# Add title
plt.title("Subscriber Counts Non Lockdown, Early Lockdown & Late Lockdown", fontsize=12)

# Manually set x-axis label
plt.xlabel("Subscriber counts")


# Show the graph
plt.savefig(output_path+'sub_count_nld_eld_lld.png', dpi=300, facecolor = 'white')
```
