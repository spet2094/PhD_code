```python
# Create a grid of 2x3 subplots to display the 6 plots
folder_path = path  

# Use glob to find all files with the specified prefix
file_pattern = ["Alywin_transport_worker_WC.png",
                "Brighton_transport_worker_WC.png", 
                "Edith_farmer_WC.png",
                "Happymore_bricklayer_WC.png",
                "Janet_ag_worker_WC.png",
                "Kennedy_transport_WC.png",
                "Munyaradzi_transport_worker_WC.png",
                "Peter_HCW_WC.png",
                "Polite_Vendor_WC.png"]

# make all files into a list of strings that can be recognised 
all_files = []

# Loop over each file name in the file pattern and use os.path.join to get the full path
for file_name in file_pattern:
    full_path = os.path.join(folder_path, file_name)
    all_files.append(full_path)

# Example usage of `all_files`
for image in all_files:
    print(image)
    
# # change the order of display 
# desired_order = [
# #     f"{folder_path}perfMix_{file_suffix_a}.png",
#     f"{folder_path}bubblesNorm_{file_suffix_a_i}.png",
#     f"{folder_path}workToHome_{file_suffix_a_i}.png",
#     f"{folder_path}comWorkersToHome_{file_suffix_a_i}.png",
#     f"{folder_path}schoolToCom_{file_suffix_a_i}.png",
#     f"{folder_path}schoolToHome_{file_suffix_a_i}.png",
#     f"{folder_path}allToHome_{file_suffix_a_i}.png"
# ]

# # # Sort the images list based on the desired order
# ordered_images = sorted(images, key=lambda x: desired_order.index(x))
```

```python
# Define your custom labels
custom_labels = [
    "Transport Worker 1",
    "Transport Worker 2",
    "Farmer",
    "Bricklayer",
    "Agricultural Worker",
    "Transport Worker 3",
    "Transport Worker 4",
    "Healthcare Worker",
    "Vendor"
]


fig, axs = plt.subplots(5, 2, figsize=(70, 50))

for ax, img, label in zip(axs.flatten(), all_files, custom_labels):
    image = mpimg.imread(img)
    ax.imshow(image)
    ax.axis('off')
    
    # Add a manually defined label to each subplot
    ax.set_title(label, fontsize=30)  # You can adjust the font size
 
# If there are more subplots than images, turn off the remaining subplots
for ax in axs.flatten()[len(all_files):]:
    ax.axis('off')
    
#plt.tight_layout(pad=0)

# Adjust the spacing: left, right, bottom, top, wspace, hspace
plt.subplots_adjust(left=0, right=0.25, top=0.5, bottom=0, wspace=0, hspace=0.05)

# Add a title for the entire figure
plt.suptitle('Wordcloud Plots for Different Workers (Manicaland Qualitative: July 2023)', fontsize=30, weight='bold', y=0.52, x=0.12)

# selective lockdown plots
plt.savefig(path+"9_wordclouds.png", dpi=150, bbox_inches='tight')
plt.show()
```
